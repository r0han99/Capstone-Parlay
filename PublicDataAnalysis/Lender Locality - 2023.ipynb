```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np

plt.rcParams['font.family'] = 'avenir'
```

```python
cleaned_data = pd.read_csv("Cleaned_fy2023.csv").drop("Unnamed: 0",axis=1)
cleaned_data
```

```python
range_ = cleaned_data['GrossApproval'].max() - cleaned_data['GrossApproval'].min()
```

```python
cleaned_data['GrossApproval'].min()
```

```python
range_
```

```python
# Define bin edges
bin_edges = list(range(0, 5000001, 250000))  
bin_labels = [f"{edge//1000}K-{(edge+250000)//1000}K" for edge in bin_edges[:-1]]
cleaned_data['Bin'] = pd.cut(cleaned_data['GrossApproval'], bins=bin_edges, labels=bin_labels, include_lowest=True)
```

```python
cleaned_data
```

## Distribution of same-city borrower and lenders pairs over various loan amount ranges

```python
same_city = cleaned_data[cleaned_data["BorrCity"]==cleaned_data['BankCity']][['BankCity','GrossApproval',"Bin"]]
same_state = cleaned_data[cleaned_data["BorrState"]==cleaned_data['BankState']][['BankState','GrossApproval',"Bin"]]
```

```python
prop_same_city = pd.DataFrame(np.round(same_city['Bin'].value_counts()/same_city.shape[0]*100,2)).reset_index()
prop_same_state = pd.DataFrame(np.round(same_state['Bin'].value_counts()/same_state.shape[0]*100,2)).reset_index()
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=prop_same_city, x='Bin', y='count', palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Gross Approved Loan Amount')
plt.ylabel('Percentage')
plt.title('Distribution of Borrowers & Lenders within same City over Loan Amount for FY2023', size=18)

# Add numbers on top of each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height()}%', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.tight_layout()
plt.show()
```

```python
# Line Graph
plt.figure(figsize=(14,7))
plt.plot(prop_same_city_within_bin['Bin'], prop_same_city_within_bin['Percentage'], marker='o', linestyle='-', color='blue')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.xlabel('Loan Amount')
plt.ylabel('Percentage of Same City Loans')
plt.title('Trend of Same City Loans Percentage by Loan Amount')
plt.grid(True)
plt.show()
```
