```python
import pandas as pd
import numpy as np
```

```python
X_train = pd.read_csv("XC_train.csv").drop('Unnamed: 0',axis=1)
X_test = pd.read_csv("XC_test.csv").drop('Unnamed: 0',axis=1)
X_val = pd.read_csv("XC_val.csv").drop('Unnamed: 0',axis=1)
```

```python
y_train = X_train['lumos_score']
y_test = X_test['lumos_score']
y_val = X_val['lumos_score']

# remove target 
X_train = X_train.drop("lumos_score",axis=1)
X_test = X_test.drop("lumos_score",axis=1)
X_val = X_val.drop("lumos_score",axis=1)
```

### Random Forest with GridSearch

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
```

```python
x_sample = X_train.sample(1000)
y_sample = y_train[x_sample.index]
```

### Running GridSearch on a Sample to Get the Best Estimators for the Data

```python

rf = RandomForestRegressor()

# Leveraging GridSearchCV to Achieve a FineTuned Model
parameters = {
    'n_estimators': [100, 200, 300],
    "max_depth": [None, 10, 20, 30],
    "min_samples_split": [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Perform grid search
grid_search = GridSearchCV(estimator=rf, param_grid=parameters, cv=5, scoring="neg_mean_squared_error")
grid_search.fit(x_sample, y_sample)

# get best parameters and best score
best_params = grid_search.best_params_
best_score = np.sqrt(-grid_search.best_score_)

print("Best Parameters:",best_params)
print("Best RMSE: ", best_score)
```

```python
# Define the best parameters
best_params = {'max_depth': 20, 
               'min_samples_leaf': 2, 
               'min_samples_split': 5, 
               'n_estimators': 200}
# Create the Random Forest model with the best parameters
rf_model = RandomForestRegressor(
    max_depth=best_params['max_depth'],
    min_samples_leaf=best_params['min_samples_leaf'],
    min_samples_split=best_params['min_samples_split'],
    n_estimators=best_params['n_estimators'],
    random_state=42  # You can set a random state for reproducibility
)

# Fit the model on the training data
rf_model.fit(X_train, y_train)
```

```python
predictions_train = rf_model.predict(X_train)
predictions_test = rf_model.predict(X_test)
predictions_val = rf_model.predict(X_val)
```

```python
predictions_train
```

```python
predictions_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Calculate Mean Absolute Error (MAE) for training set
mae_train = mean_absolute_error(y_train, predictions_train)

# Calculate Mean Squared Error (MSE) for training set
mse_train = mean_squared_error(y_train, predictions_train)

# Calculate R-squared (R2) score for training set
r2_train = r2_score(y_train, predictions_train)

# Calculate Mean Absolute Error (MAE) for test set
mae_test = mean_absolute_error(y_test, predictions_test)

# Calculate Mean Squared Error (MSE) for test set
mse_test = mean_squared_error(y_test, predictions_test)

# Calculate R-squared (R2) score for test set
r2_test = r2_score(y_test, predictions_test)

# Calculate Mean Absolute Error (MAE) for validation set
mae_val = mean_absolute_error(y_val, predictions_val)

# Calculate Mean Squared Error (MSE) for validation set
mse_val = mean_squared_error(y_val, predictions_val)

# Calculate R-squared (R2) score for validation set
r2_val = r2_score(y_val, predictions_val)

print("Training Set Metrics:")
print("Mean Absolute Error (MAE):", mae_train)
print("Mean Squared Error (MSE):", mse_train)
print("R-squared (R2) Score:", r2_train)
print()
print("Test Set Metrics:")
print("Mean Absolute Error (MAE):", mae_test)
print("Mean Squared Error (MSE):", mse_test)
print("R-squared (R2) Score:", r2_test)
print()
print("Validation Set Metrics:")
print("Mean Absolute Error (MAE):", mae_val)
print("Mean Squared Error (MSE):", mse_val)
print("R-squared (R2) Score:", r2_val)
```

```python
import pandas as pd

# Example function to generate recommendations
def generate_recommendations(sample_df, model, sample):
    # Convert sample dictionary to DataFrame
    sample_df = pd.DataFrame(sample_df)

    # Predict score for the sample
    predicted_score = model.predict(sample_df)[0]
    print("Predicted score:", predicted_score)

    # Extract the columns for recommendation
    recommendation_cols = ['rate_type', 'businesstype', 'franchise', 'borr_count',
                           'loan_max_fico', 'loan_min_fico', 'loan_mean_fico', 'loan_size']

    print("\nRecommendations:")
    for col in recommendation_cols:
        #print(sample)
        original_value = sample[col]
        
        # Example recommendation: Increase loan_max_fico
        modified_sample = sample.copy()
        
        # Modify feature value (e.g., increase by 10%)
        modified_sample[col] += 10 if isinstance(original_value, int) or isinstance(original_value, float) else 1
        
        # Predict score with modified sample
        modified_score = model.predict(pd.DataFrame([modified_sample]))[0]
        
        # Compare scores
        if modified_score > predicted_score:
            print(f"Change {col} to improve score (predicted score: {modified_score})")

# Sample data
sample_data = {'probability_of_default': 1.83,
 'loss_given_default': 76.37,
 'expected_loss': 1.39,
 'rate_type': 1,
 'franchise': False,
 'businesstype': 0,
 'business_type': 0,
 'empoyessexistingqty': 4,
 'borr_count': 1,
 'loan_max_fico': 661,
 'loan_min_fico': 661,
 'loan_mean_fico': 661.0,
 'collateral': True,
 'primary_collateral_type': 2,
 'collateral_amount': 2,
 'loan_size': 1,
 'business_revenue': 741000,
 'bankruptcy': 0,
 'lawsuit': 0,
 'current_inflation_rate': 1.73,
 'expected_12mth_inflation_rate': 3.0,
 'fed_funds': 0.09,
 't10yie': 2.13,
 'vix': 17.08,
 'vix_change': 7.62,
 'avg_hourly_earnings': 1.96,
 't10y2y': 1.34,
 'county_unemployment_rate': 7.9,
 'state_unemployment_rate': 13.7,
 'county_unemployment': 2260.0,
 'state_unemployment': 163681.0,
 'county_employment': 26410.0,
 'state_employment': 1030775.0,
 'county_labor_force': 28670.0,
 'state_labor_force': 1194456.0,
 'state_civilian_labor_population': 4591500.0,
 'state_employment_pop_ratio': 61.5,
 'state_labor_force_participation_rate': 66.6,
 'Risk_Group': 4
}



# Generate recommendations
generate_recommendations(pd.DataFrame(pd.Series(sample_data)).T, rf_model, sample_data)
```
